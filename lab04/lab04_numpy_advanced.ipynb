```python
##zad 2a
# Dane miesięczne z jednego roku (flat)
dane_flat = np.array([120, 135, 98, 142, 167, 189, 201, 178, 156, 143, 198, 221])
# Przekształć na kwartały (4 kwartały × 3 miesiące)
kwartaly = dane_flat.reshape(4, 3)
print(f"Kwartały:\n{kwartaly}")
print(f"Suma per kwartał: {kwartaly.sum(axis=1)}")
print(f"Najlepszy kwartał: Q{kwartaly.sum(axis=1).argmax() + 1}")
```

```python
## zad 2b
# Dane z 3 oddziałów (osobne tablice)
warszawa = np.array([150, 160, 170, 180])
krakow = np.array([120, 130, 140, 150])
wroclaw = np.array([90, 100, 110, 120])
# Połącz w macierz (3 oddziały × 4 kwartały)
wszystkie = np.vstack([warszawa, krakow, wroclaw])
print(f"Wszystkie:\n{wszystkie}")
print(f"Suma per oddział: {wszystkie.sum(axis=1)}")
print(f"Suma per kwartał: {wszystkie.sum(axis=0)}")
```

```python
##Zad 2c 
# Masz dane o sprzedaży 6 produktów jako wektor
sprzedaz_flat = np.array([100, 200, 150, 300, 250, 180, 120, 220, 170, 310, 260, 190])
# Wiesz, że to 6 produktów × 2 miesiące
# 1. Reshape na macierz 6×2
macierz = sprzedaz_flat.reshape(6, 2)
# Podpowiedź: sprzedaz_flat.reshape(6, 2) albo .reshape(-1, 2)
# 2. Oblicz średnią sprzedaż per produkt (axis=1)
srednia = macierz.mean(axis=1)
# 3. Oblicz zmianę między miesiącem 1 a 2 (kolumna 1 - kolumna 0)
# Podpowiedź: macierz[:, 1] - macierz[:, 0]
zmiana = macierz[:, 1] - macierz[:, 0]

print("Macierz:\n", macierz)
print("Średnia:", srednia)
print("Zmiana:", zmiana)
```

# ZADANIE 3

```python
## Dane 
np.random.seed(2026)
# 8 pracowników, dane roczne
pracownicy = ['Anna', 'Jan', 'Ewa', 'Marek', 'Kasia', 'Piotr', 'Zofia', 'Tomek']
pensje = np.array([5500, 7200, 4800, 9100, 6300, 4200, 8500, 5800])
staz = np.array([3, 8, 2, 15, 5, 1, 12, 4])
oceny = np.array([4.2, 3.8, 4.5, 4.0, 4.7, 3.5, 4.3, 3.9])
# Sprzedaż miesięczna 4 produktów przez 6 miesięcy
produkty_nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Akcesorium']
sprzedaz = np.random.randint(20, 200, size=(4, 6))
ceny_produktow = np.array([3500, 1800, 2500, 150])
```

```python
#Zad 3a
print("Średnia:", pensje.mean())
print("Mediana:", np.median(pensje))
print("Odchylenie std:", pensje.std())

nazwy = np.array(pracownicy)

print("Powyżej średniej:", nazwy[pensje > pensje.mean()])
print("Poniżej mediany:", nazwy[pensje < np.median(pensje)])

rozstep = pensje.max() - pensje.min()
iqr = np.percentile(pensje, 75) - np.percentile(pensje, 25)

print("Rozstęp:", rozstep)
print("IQR:", iqr)
```

```python
##Zad 3b
premia = np.where(oceny >= 4.5, 2000,
         np.where(oceny >= 4.0, 1000, 0))

nowe_pensje = pensje + premia

print("Premie:", premia)
print("Nowe pensje:", nowe_pensje)

print("Najwyższa premia:", pracownicy[premia.argmax()])
```

```python
##Zad 3c 
kolejnosc = np.argsort(pensje)[::-1]

for i, idx in enumerate(kolejnosc, 1):
    print(i, pracownicy[idx], pensje[idx], "staż:", staz[idx])
```

```python
## Zad 3d
korelacja_staz_pensja = np.corrcoef(staz, pensje)[0, 1]
korelacja_staz_ocena = np.corrcoef(staz, oceny)[0, 1]

print("Staż vs pensja:", korelacja_staz_pensja)
print("Staż vs ocena:", korelacja_staz_ocena)
```

```python
## Zad 3e 
ceny_kol = ceny_produktow.reshape(4, 1)
przychod = sprzedaz * ceny_kol

# suma per produkt
suma_produkty = przychod.sum(axis=1)
print("Najlepszy produkt:", produkty_nazwy[suma_produkty.argmax()])

# suma per miesiąc
suma_miesiace = przychod.sum(axis=0)
print("Najlepszy miesiąc:", suma_miesiace.argmax() + 1)

# trend
pierwsze = przychod[:, :3].sum()
ostatnie = przychod[:, 3:].sum()

print("Pierwsze 3 miesiące:", pierwsze)
print("Ostatnie 3 miesiące:", ostatnie)
```

```python
##Krok 1

import numpy as np

np.random.seed(42)

wiek = np.clip(np.random.normal(35, 10, 100).astype(int), 18, None)
wydatki = np.clip(np.random.normal(500, 200, 100), 0, None)
wizyty = np.random.poisson(5, 100)
```

```python
##Krok 2
# 1. Statystyki opisowe każdej zmiennej (mean, median, std, min, max)
# Podpowiedź: print(f"Wiek — średnia: {wiek.mean():.1f}, mediana: {np.median(wiek):.1f}")
print(f"Wiek — średnia: {wiek.mean():.1f}, mediana: {np.median(wiek):.1f}, std: {wiek.std():.1f}, min: {wiek.min()}, max: {wiek.max()}")

print(f"Wydatki — średnia: {wydatki.mean():.1f}, mediana: {np.median(wydatki):.1f}, std: {wydatki.std():.1f}, min: {wydatki.min():.1f}, max: {wydatki.max():.1f}")

print(f"Wizyty — średnia: {wizyty.mean():.1f}, mediana: {np.median(wizyty):.1f}, std: {wizyty.std():.1f}, min: {wizyty.min()}, max: {wizyty.max()}")
```

```python
# 2. Korelacja wiek vs wydatki
# Podpowiedź: np.corrcoef(wiek, wydatki)[0, 1]
korelacja_wiek_wydatki = np.corrcoef(wiek, wydatki)[0, 1]
print("Korelacja wiek vs wydatki:", korelacja_wiek_wydatki)
```

```python
# 3. Korelacja wizyty vs wydatki
korelacja_wizyty_wydatki = np.corrcoef(wizyty, wydatki)[0, 1]
print("Korelacja wizyty vs wydatki:", korelacja_wizyty_wydatki)
```

```python
# 4. Segmentacja klientów za pomocą np.where:
# - "WYSOKI" gdy wydatki > średnia + std
# - "NISKI" gdy wydatki < średnia - std
# - "ŚREDNI" w pozostałych przypadkach
# Podpowiedź: zagnieżdżony np.where (jak w zadaniu 3b)
# sr = wydatki.mean()
# sd = wydatki.std()
# segment = np.where(wydatki > sr + sd, 'WYSOKI', np.where(wydatki < sr - sd, 'NISKI', 'ŚREDNI'))
sr = wydatki.mean()
sd = wydatki.std()

segment = np.where(
    wydatki > sr + sd, "WYSOKI",
    np.where(wydatki < sr - sd, "NISKI", "ŚREDNI")
)

print(segment)
```

```python
# 5. Ile klientów w każdym segmencie? Użyj np.unique(..., return_counts=True)
unikalne, liczby = np.unique(segment, return_counts=True)

for s, l in zip(unikalne, liczby):
    print(s, ":", l)
```
